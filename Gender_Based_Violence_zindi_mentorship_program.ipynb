code of bbx plot inspired from this github
https://github.com/AarohiSingla/Faster-R-CNN/blob/main/data_prep.ipynb

```
def plot_bbox(img_id):
  img_id=str(img_id).zfill(4)+'.jpg'
  #img_url = data_train.loc[data_train["iamge_id"]==img_id]['OriginalURL'].values[0]
  img_path = '/content/drive/MyDrive/Datasets/Train/images/'+ str(data_train.loc[data_train["image_id"]==img_id].values[0][0])
  img = io.imread(img_path)
  height, width, channel = img.shape
  print(f"Image: {img.shape}")
  bboxs = data_train.loc[data_train["image_id"]==img_id]
  for index, row in bboxs.iterrows():
      xmin = row['Xmin']
      xmax = row['Xmax']
      ymin = row['Ymin']
      ymax = row['Ymax']
      xmin = int(xmin*width)
      xmax = int(xmax*width)
      ymin = int(ymin*height)
      ymax = int(ymax*height)
      class_name = bboxs.values[0][5] 
      #class_series = class_descriptions[class_descriptions[0]==label_name]
      #class_name = class_series[1].values[0]
      print(f"Coordinates: {xmin,ymin}, {xmax,ymax}")
      cv2.rectangle(img, (xmin,ymin), (xmax,ymax), (255,0,0), 5)
      font = cv2.FONT_HERSHEY_SIMPLEX
      cv2.putText(img, class_name, (xmin,ymin-10), font, 1, (0,255,0), 2)
  plt.figure(figsize=(15,10))
  plt.title('Image with Bounding Box')
  plt.imshow(img)
  plt.axis("off")
  plt.show()
```
